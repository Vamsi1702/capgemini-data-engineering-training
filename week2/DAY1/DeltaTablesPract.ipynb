```sql
create database mydeltadb;
```

```sql
use mydeltadb;
```

```sql
drop table if exists firsttable;
```

```sql
create table firsttable(
  id int,
  name string,
  class string,
  age int,
  gender string
)using delta;
```

```sql
show tables;
```

```sql
describe formatted firsttable;
```

```sql
insert into firsttable (id,name,gender)values(6,'vvv','male');
```

```sql
select * from firsttable;
```

```sql
select * from firsttable
where class is null;
```

```sql
update firsttable
set name='kkk',age=25
where id=3;
```

```sql
delete from firsttable
where class='second'
```

```sql
truncate table firsttable;
```

```sql
create table stagingtable2(
  id int,
  name string,
  class string,
  age int,
  gender string,
  action string
)using delta;
```

```sql
insert into stagingtable2(id,name,class,age,gender,action)values(1,'vvv','first',20,'male','u'),(2,'vvv','first',35,'male','i'),(3,'vvv','first',20,'male','d');
```

```sql
select * from firsttable;
```

```sql
insert into firsttable
select * from stagingtable;
```

```sql
truncate table stagingtable2;
```

```sql
select * from stagingtable2;
```

```sql
merge into firsttable
using stagingtable2
on firsttable.id = stagingtable2.id
when matched and stagingtable2.action='u' then update set *
when matched and stagingtable2.action='d' then delete
when not matched and stagingtable2.action='i' then insert *;
```
