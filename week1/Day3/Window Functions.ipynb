```sql
select emp_name,salary,
row_number() over(order by salary desc) as RowNum
 from emp1234
```

```sql
select emp_name,salary,department,
row_number() over(partition by department order by salary desc) as RowNum
 from emp1234
```

```sql
select emp_name,salary,join_date,
row_number() over(order by join_date desc) as JoinDate
 from emp1234
```

```sql
select department,salary,emp_name,join_date,
row_number() over(partition by department order by join_date asc) as rn
from emp1234
```

```sql
select emp_name,department,salary,
row_number() over(order by salary asc) as rn
from emp1234
```

```sql
select emp_name,department,
row_number() over(order by emp_name asc) as rn
from emp1234
```

**Rank**

```sql
select emp_name,department,salary,
rank() over(order by salary desc) as rn
from emp1234
```

```sql
select emp_name,department,salary,
rank() over(partition by department order by salary desc) as rn
from emp1234
```

```sql
select emp_name,salary,join_date,
rank() over(order by join_date desc) as JoinDate
 from emp1234
```

**Dense Rank**
